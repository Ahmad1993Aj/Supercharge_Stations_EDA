```python
df.drop("Unnamed: 11", axis=1, inplace=True)
```

```python
df.head(2)
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,round(pct_missing*100)))
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
#### Eine Methode für die Be handung mit Messing-Data:
```

```python
df['State'].fillna(df['State'].mode()[0], inplace=True)
```

```python
mean_kw = df['kW'].mean()
df['kW'].fillna(mean_kw, inplace=True)
```

```python
df['Open Date'].fillna(df['Open Date'].mode()[0], inplace=True)
```

```python
df['Zip'].fillna(df['Zip'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# Change the data type of the Open Data column
df['Open Date']= pd.to_datetime(df['Open Date'])
```

```python
# Now let's have look at the table.
df.sort_values(by=['Open Date'], inplace=False, ascending=True)
```

```python
### andere Preprocessing methode:
```

```python
data= pd.read_csv("Supercharge Locations.csv", encoding='latin-1')
data.head()
```

```python
data.drop("Unnamed: 11", axis=1, inplace=True)
```

```python
data['State'] = data['State'].map(lambda x: x if pd.notnull(x) else 'Unknown')
```

```python
data['Zip'] = data['Zip'].map(lambda x: x if pd.notnull(x) else -1)
```

```python
avg_kw = data['kW'].mean()
data['kW'] = data['kW'].map(lambda x: x if pd.notnull(x) else avg_kw)
```

```python
data['Open Date'].fillna(data['Open Date'].mode()[0], inplace=True)
```

```python
data.head()
```

```python
sns.pairplot(data);
```
